Notebook for creating text files from excel sheets containing article text.

Pay attention to encoding - trying to do word counting.

Create two text files for each month:
* quotes: 2023_01_jan_CBC_news_sentiment_quote_text.txt
* non quotes: 2023_01_jan_CBC_news_sentiment_non_quote_text.txt

Put all these new files into a new higher level directory called 'CBC_output_corpus'

```python
import os
import re
import pandas as pd
```

```python
# helper function for individually processing each file
def to_txt(df, output_path):
  with open(output_path, 'w', encoding='utf-8') as f:
    f.write(df.to_string(header=False, index=False))
```

```python
# driver code
input_dir = '/content/drive/My Drive/evaluation_quotes/CBC_qt_output'
output_dir = '/content/drive/My Drive/evaluation_quotes/CBC_output_corpus'

xl_path = ''
txt_path = ''

for f in os.scandir(input_dir):
  if f.is_file():
    xl_path = f.path

    # get file name and produce text file names
    filename = os.path.basename(f)
    quote_fp = re.sub(r'clean_new.xlsx', 'quote_text.txt', filename)
    non_quote_fp = re.sub(r'clean_new.xlsx', 'non_quote_text.txt', filename)

    # produce output paths
    quote_output = os.path.join(output_dir, quote_fp)
    non_quote_output = os.path.join(output_dir, non_quote_fp)

    # read article text into dataframes
    quotes_df = pd.read_excel(xl_path, usecols=['quote'])
    non_quotes_df = pd.read_excel(xl_path, usecols=['non_quoted_text'])

    # pass dataframes to text file function
    to_txt(quotes_df, quote_output)
    to_txt(non_quotes_df, non_quote_output)
```
